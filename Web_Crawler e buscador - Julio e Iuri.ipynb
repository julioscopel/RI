```python
Licitacoes= pd.DataFrame(columns=['Número do Edital','Orgão','Setor Requisitante', 'Processo' , 'Modalidade', 'Pregoeiro/Presidente/Equipe', 'Resultado', 'OBJETO'])
```

```python
numeroinicial = 450 #2
numerofinal = 470 #3928

for numero in range(numeroinicial,numerofinal):
    
    try:
        r = requests.get("https://www.vilavelha.es.gov.br/licitacoes/view/" + str(numero))
        r.raise_for_status()
    except HTTPError:
        print ("Could not download page ", "https://www.vilavelha.es.gov.br/licitacoes/view/" + str(numero))
    else:
        s = r.text # retorna html
        soup = BeautifulSoup(s, 'html.parser')
        count = 0
        registro = []
        for p in soup.find_all('p'):
            if count<8:
                registro.append(p.text)
                if registro[count][:1] == '\n':
                    del registro[count]
                    count = count - 1
                count = count + 1
        if count == 8:
            #trata registros
            registro[0] = registro[0][19:]
            registro[1] = registro[1][7:]
            registro[2] = registro[2][20:]
            registro[3] = registro[3][10:]
            registro[4] = registro[4][12:]
            registro[5] = registro[5][29:]
            registro[6] = registro[6][11:]
            #Adiciona resgistros no DF Licitacao
            Licitacoes.loc[len(Licitacoes)] = registro
```

```python
#Higienizando DataSet
docs_cleaned = []

for i in range(len(Licitacoes['OBJETO'])):
    docs_cleaned.append(Licitacoes['OBJETO'][i].lower().replace(".", "").replace(",", "").replace("\n","").replace(":","").replace("\t",""))
```

```python
vocabulary = {}
'''Inicializa o vocabulário'''
for i in range(len(docs_cleaned)):
    for j in docs_cleaned[i].split(" "):
        vocabulary[j] = 0

'''Contabiliza as ocorrências dos documentos'''
for word in vocabulary.keys():
    for doc in docs_cleaned:
        if word in doc.split():
            vocabulary[word] += 1
        
print(vocabulary)

invertedList = dict()
'''Contabiliza a frequência dos termos nos documentos'''
for term in vocabulary.keys():
    invertedList[term] = list()
    indiceDocumento = 0
    for doc in docs_cleaned:
        frequencia = 0
        for word in doc.split():
            if word == term:
                frequencia += 1
        if frequencia > 0:
            invertedList[term].append([indiceDocumento, frequencia])
        indiceDocumento += 1

print(invertedList)
```

```python
#Exibe os dados formatados
fmt = '{:<8}{:<15}{:<8}{}'

print(fmt.format('', 'Vocabulary', 'ni', 'Inverted Lists'))
for i, (vocab, ni, ocurrencies) in enumerate(zip(vocabulary.keys(), vocabulary.values(), invertedList.values())):
    print(fmt.format(i, vocab, ni, ocurrencies))
```

```python
simpleSearchTerm = input("Digite o termo de busca: ")
if simpleSearchTerm.lower() in vocabulary.keys():
    print("Foram encontrados ", vocabulary.get(simpleSearchTerm.lower()), " documentos com o termo de busca...")
    
    print()
    for i in invertedList.get(simpleSearchTerm.lower()):
        term = simpleSearchTerm.lower()
                                                         #https://en.wikipedia.org/wiki/ANSI_escape_code
                                                         #256-color mode — foreground: ESC[38;5;#m   background: ESC[48;5;#m
        print((i[0]), " - ", Licitacoes['Número do Edital'][i[0]], "-", docs_cleaned[i[0]].replace(term, "\033[48;5;0m\033[38;5;226m {term} \033[0;0m".format(term=term)))
        
else:
    print("Termo não encontrado!")

```

```python
#Condição para uso de AND ou OR
while True:
  tipodebusca = input("Informe o tipo de busca complexa (And ou Or): ")
  if not re.match("^[and,or]*$", tipodebusca.lower()):
    print ("Erro! Somente And ou Or como entrada")
  else:
    break

#complexSearchTerms = exs: "prefeitura de" "aquisição  de"
complexSearchTerms = input("Digite os termos de busca: ")
termsQueries = complexSearchTerms.split()

resultListPerTerm = {}
resultList = []

#Monta as listas resultantes da consulta com N termos
for term in termsQueries:
    lista = []
    
    if invertedList.get(term.lower()) != None:
        for i in invertedList.get(term.lower()):
            lista.append(i[0])
        resultListPerTerm[term] = lista
        resultList.append(lista)
    
#Lista com o
for i, (k, v) in enumerate(resultListPerTerm.items()):
    print("{:<8}key: {:<20} value: {}".format(i, k, v))

print()
    
responseSet = []
#Realiza a interseção entre os conjuntos
for i in range(len(resultList)-1):
    if tipodebusca.lower() == 'and':
    #Operador AND
        responseSet.append(list(set(resultList[i]).intersection(resultList[i+1])))
    #Operador OR
    else:
        responseSet.append(list(set(resultList[i]).union(resultList[i+1])))

print("Foram encontrados ", len(np.unique(responseSet)), " documentos com o termo de busca...")
print()

idf = 1 + math.log(len(docs_cleaned)/(len(np.unique(responseSet))))

print ('idf = ', idf )

'''Cria variações com capitalize e uppercase para marcar o texto'''
copiesTermsQueries = termsQueries.copy()
([termsQueries.append(word.capitalize()) for word in copiesTermsQueries])
([termsQueries.append(word.upper()) for word in copiesTermsQueries])

#Monta o Resultado
lista = []
for doc in np.unique(responseSet):
    documento = docs_cleaned[doc]
    for term in termsQueries:
        documento = documento.replace(term, "\033[48;5;0m\033[38;5;226m {term} \033[0;0m".format(term=term))        
    lista.append(str(str(doc+1) + " - " + documento))
    
#Exibe o Resultado
for resultado in lista:
    print(resultado)
```
